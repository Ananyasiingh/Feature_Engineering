Feature Engineering using percentile 


```python
import pandas as pd

df = pd.read_csv("heights.csv")
```

```python
df.head()
```

```python
max_thresold=df['height'].quantile(0.95)
```

This means that 95%ile is 9.68999 and above this will be considered as an outlier.

```python
df[df['height']>max_thresold]
```

```python
min_thresold = df['height'].quantile(0.05)
min_thresold
```

```python
df[df['height']<min_thresold]
```

```python
df[(df['height']<max_thresold) & (df['height']>min_thresold)]
```

Here, the 75% of 'Total_sqft' is less than 1672.00(this is how we read this).

```python
#we can also apply an array function to repeat the above function
min_thresold, max_thresold = df.price_per_sqft.quantile([0.001, 0.999])
min_thresold, max_thresold
```

```python
df2 = df[(df['price_per_sqft']>min_thresold) & (df['price_per_sqft']<max_thresold)]
df2
```
